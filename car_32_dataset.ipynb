```python
import pandas as pd
import numpy as np 
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.neighbors import KNeighborsClassifier  
from sklearn.ensemble import RandomForestClassifier
```

```python
link_for_dataset = 'https://code.datasciencedojo.com/datasciencedojo/datasets/repository/master/archive.zip'
df = pd.read_csv('/content/car.data')
df.head()
```

```python
df.isna().sum()
```

```python
encoder = LabelEncoder()
df['vhigh'] = encoder.fit_transform(df['vhigh'])
```

```python
df.head()
```

```python
df['vhigh.1'] = encoder.fit_transform(df['vhigh.1'])
df['small'] = encoder.fit_transform(df['small'])
df['low'] = encoder.fit_transform(df['low'])
df.head()
```

```python
df.replace('5more', 5, inplace = True)
df.replace('more', 5, inplace = True)
df['2.1'].value_counts()
```

```python
df['2.1'].value_counts()
```

```python
X = df.drop(columns = 'unacc')
Y = df['unacc']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, random_state = 1, test_size = 0.2)
```

Now we will train our model with model Logistic Regression


```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
pre = model.predict(x_test)
pre2 = model.predict(x_train)
```

As you can see accuracy is low, so we will train our data with another model

```python
print(classification_report(pre2, y_train))
```

This model is KNeighborsClassifier.At first in this model we should find value of K

K equals 9 


```python
from sklearn.neighbors import KNeighborsClassifier  
for i in range(1,40):
  model2 = KNeighborsClassifier(n_neighbors=i)
  model2.fit(x_train, y_train)
  pre3 = model2.predict(x_test)
  print(f"__________________________________________________________",i)
  print(classification_report(pre3,y_test))
```

As you can see accuracy is 95. It is the high result


```python
# for i in range(1,50):
model3 = RandomForestClassifier(n_estimators=11)
model3.fit(x_train, y_train)
pre4 = model3.predict(x_test)
print('___________________________________________',i)
print(classification_report(pre4, y_test))
```
